### DATA MANIPULATION

```python
# Dropping unwanted columns in a dataset
df.drop(columns=['Unnamed: 0'],inplace=True)
```

```python
missing_values=df.isnull().sum()
print("Missing Values:\n",missing_values)
```

```python
df.dropna(inplace=True)
# Dropped missing/empty cells in a column
```

```python
missing_values=df.isnull().sum()
print("Missing Values:\n",missing_values)
```

```python
df.describe()
# knowing how the data will look like mathematically
```

```python
# categorical columns is not a built-in function in python but it is a concept commonly used in data manipulation 
# df.select_dtypes(include=['object']).columns : to select columns whose dtype is an object not a float or an int
#unique_values[column]=df.columns.unique(): select unique values in the already selected columns
# df.transpose: changes the shape of a dataset , arranging rows and columns in the proper format
# orient='index': is used to arrange the index of the dataframe


unique_values={}
categorical_columns=df.select_dtypes(include=['object']).columns
for columns in categorical_columns:
    unique_values[columns]=df[columns].unique()
    unique_values_df=pd.DataFrame.from_dict(unique_values,orient='index')
    unique_values_df=unique_values_df.transpose()
print("Unique Values for Categorical Columns:")
unique_values_df.head()

```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')
sns.set_palette("deep")
```

```python
fig,axes= plt.subplots(2,2, figsize=(12,10))
# plot 1: Histogram of movie ratings
sns.histplot(data=df,x='Movie_Rating',ax=axes[0,0],bins=20,kde=True)
axes[0,0].set_title('Distribution of Movie Rating')
axes[0,0].set_xlabel('Movie Rating')
axes[0,0].set_ylabel('Count')

#plot2: Boxplot of number of ratings
sns.boxplot(data=df,y='No_of_Ratings',ax=axes[0,1],showfliers=False)
axes[0,1].set_title('Boxplot Number of Ratings')
axes[0,1].set_ylabel('Number of Ratings')

#plot3: Barplot of MPAA Ratings
sns.countplot(data=df,x='MPAA_Rating',ax=axes[1,0])
axes[1,0].set_title('Distribution of MPAA Ratings')
axes[1,0].set_xlabel('MPAA Rating')
axes[1,0].set_ylabel('Count')

#plot4: Scatter plot of ReleaseYear vs Price
sns.scatterplot(data=df,x='ReleaseYear',y='Price', ax=axes[1,1])
axes[1,1].set_title('Scatterplot of ReleaseYear vs Price')
axes[1,1].set_xlabel('ReleaseYear')
axes[1,1].set_ylabel('Price')

plt.suptitle('VISUAL FOR AMAZON MOVIE DATASET', fontsize=16)
plt.tight_layout(rect=[0,0,1,0.95])

plt.show()
```
